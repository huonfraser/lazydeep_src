```python
def batch_predict(models,X,y):
    preds = {name:[] for name in models.keys()}
    for xi, yi in stream.iter_pandas(X,y):
        for name,model in river_models.items():
            pred = model.predict_one(xi)
            preds[name].append(pred)
    return preds
        
def batch_score(models,X,y):
    preds = {name:[] for name in models.keys()}
    for xi, yi in stream.iter_pandas(X,y):
        for name,model in river_models.items():
            pred = model.predict_one(xi)
            preds[name].append(pred)
            
    scores = {name:r2_score(y,pred) for name,pred in preds.items()}
    mse = {name:mean_squared_error(y,pred) for name,pred in preds.items()}
    return scores, mse
        

def batch_learn(models,X,y):
    for xi, yi in stream.iter_pandas(X,y):
        for name,model in river_models.items():
             model.learn_one(xi,yi)
    return models
```

```python
def sk_predict_one(models,xi,yi):
    preds = {}
    for name,model in models.items():
        preds[name] = models.predict([dict2numpy(xi)])[0]
    
    return preds  

def sk_predict(models,X,y):
    preds = {}
    for name,model in models.items():
            pred = model.predict(X)
            preds[name]=pred.tolist()
    return preds
        
def sk_score(models,X,y):
    preds = {}
    for name,model in models.items():
        pred = model.predict(X)
        preds[name]=pred
            
    scores = {name:r2_score(y,pred) for name,pred in preds.items()}
    mse = {name:mean_squared_error(y,pred) for name,pred in preds.items()}
    return scores, mse
        

def sk_learn(models,X,y):
    for name,model in models.items():
        model.fit(X,y)
    return models
```

```python
data = pd.read_csv(data_file)
data=data.sample(frac=1,random_state=random_state)

pre_ind =[i for i in range(0,10000)]
pretrain_ind,pretest_ind = train_test_split(pre_ind,train_size=5/6,random_state=random_state,shuffle=False)
stream_ind = [i for i in range(10000,110000)]


pretrain_data =  ut.TabularDataset(data.iloc[pretrain_ind,:],id_cols = id_cols, cat_cols=None, output_cols=output_cols, ignore_cols= None)
pretest_data = ut.TabularDataset(data.iloc[pretest_ind,:],id_cols = id_cols, cat_cols=None, output_cols=output_cols, ignore_cols= None)
stream_data = ut.TabularDataset(data.iloc[stream_ind,:],id_cols = id_cols, cat_cols=None, output_cols=output_cols, ignore_cols= None)

nrow, ncol = data.shape
nrow_train = len(pretrain_data)
nrow_test = len(pretest_data)
nrow_stream = len(stream_data)

print(f"train: {nrow_train}, test: {nrow_test}, stream: {nrow_stream}")
```

```python
#setup evaluation

def setup_models():
    return {'lr':LinearRidge(),
            'plsr':PLSRegression(n_components=25),
            'lwr':LocalWeightedRegression(n_neighbours=800,normalize=True),
            'pls_lwr':PLSLWR(n_components=25, n_neighbours=300)
           }

#pls-deep - random_82, => random 24, random_10
#pls-deep lwr - random_82 - lwr_k=1000 -

#deep random 29 => random 60 => random 63
#deep-lwr - random_13 - lwr_k=800

deep_model_dir = Path("D:/workspace/lazydeep/experiments/1.01/PLN7")
pls_deep_model = torch.load(deep_model_dir/"models"/"random_82"/"_model")
pls_deep_model.load_state(deep_model_dir/"models"/"random_82"/"_final")
pls_scaler = PLSRegression(n_components=34).from_state(PLSRegression(n_components=34).load_state(deep_model_dir/'preprocessing'/f"_final"))       
                          
pls_deep_lwr = LocalWeightedRegression(n_neighbours=1000,normalize=True)
                          
deep_model_dir = Path("D:/workspace/lazydeep/experiments/2.00/PLN7")
deep_model = torch.load(deep_model_dir/"models"/"random_29"/"_model")
deep_model.load_state(deep_model_dir/"models"/"random_29"/"_final")
deep_scaler = StandardScaler().from_state(StandardScaler().load_state(deep_model_dir/'preprocessing'/f"_final"))                
                      
deep_lwr = LocalWeightedRegression(n_neighbours=1000,normalize=False)

#ew have learnt - don't standardise lwr after deep
#we can view features after pls
```

```python
#setup our metrics and stores of results
river_models = setup_models()
full_set = [*river_models.keys(),*['pls_deep','pls_deep_lwr','deep','deep_lwr']]

metrics = {'R2':{name:metrics.R2() for name in full_set},
           'R2_rolling':{name:metrics.Rolling(metrics.R2(), window_size=1000) for name in full_set},
           'MSE':{name:metrics.MSE() for name in full_set},
           'MSE_rolling':{name:metrics.Rolling(metrics.MSE(), window_size=1000) for name in full_set}
          }
           
scores = {'R2':{name:[] for name in full_set},
          'MSE':{name:[] for name in full_set},
          'R2_rolling':{name:[] for name in full_set},
          'MSE_rolling':{name:[] for name in full_set}
         }    
```

```python
#test different pretraining schemes
#nrow_test

train_X,train_y = pretrain_data[:]

#train models we train
sk_learn(river_models,train_X,train_y)
#test these models 
test_X, test_y = pretest_data[:]

batch_r2,batch_mse = sk_score(river_models,test_X,test_y)
preds = sk_predict(river_models,test_X,test_y)
preds['y'] = test_y.tolist()
#add in pls-deep
#pls_scaler.fit(train_X,train_y)
test_X_t = pls_scaler.transform(test_X)
pred = pls_deep_model(torch.Tensor(test_X_t)).detach().numpy()
preds['pls_deep'] = pred.tolist()
batch_r2['pls_deep'] = r2_score(test_y,pred)
batch_mse['pls_deep'] = mean_squared_error(test_y,pred)

#add in pls-deep-lwr
train_X_t = pls_scaler.transform(train_X)
train_X_tt = pls_deep_model.compress(torch.Tensor(train_X_t)).detach().numpy()
pls_deep_lwr.fit(train_X_tt,train_y)

test_X_tt = pls_deep_model.compress(torch.Tensor(test_X_t)).detach().numpy()
pred = pls_deep_lwr.predict(test_X_tt)
preds['pls_deep_lwr'] = pred.tolist()
batch_r2['pls_deep_lwr'] = r2_score(test_y,pred)
batch_mse['pls_deep_lwr'] = mean_squared_error(test_y,pred)

#add in deep
#deep_scaler.fit(train_X,train_y)
test_X_t = deep_scaler.transform(test_X)
pred = deep_model(torch.Tensor(test_X_t)).detach().numpy()
preds['deep'] = pred.tolist()
batch_r2['deep'] = r2_score(test_y,pred)
batch_mse['deep'] = mean_squared_error(test_y,pred)

#add in deep-lwr
train_X_t = deep_scaler.transform(train_X)
train_X_tt = deep_model.compress(torch.Tensor(train_X_t)).detach().numpy()
deep_lwr.fit(train_X_tt,train_y)

test_X_tt = deep_model.compress(torch.Tensor(test_X_t)).detach().numpy()
pred = deep_lwr.predict(test_X_tt)
preds['deep_lwr'] = pred.tolist()
batch_r2['deep_lwr'] = r2_score(test_y,pred)
batch_mse['deep_lwr'] = mean_squared_error(test_y,pred)

```

```python
for name in full_set:
    model_preds = preds[name]
    for i, pred in enumerate(model_preds):
        y=test_y[i]
        #print(y,",",pred)
        for metric_k,metric_v in metrics.items():
            metric_v[name].update(y, pred)     
```

```python
for name in full_set:
    scores['R2'][name] = scores['R2'][name] + [batch_r2[name] for _ in range(0,nrow_test)]
    scores['R2_rolling'][name] = scores['R2_rolling'][name] +  [batch_r2[name] for _ in range(0,nrow_test)]
    scores['MSE'][name] = scores['MSE'][name] +  [batch_mse[name] for _ in range(0,nrow_test)]
    scores['MSE_rolling'][ name] = scores['MSE_rolling'][name] +   [batch_mse[name] for _ in range(0,nrow_test)]
```

```python
X,y = stream_data[:]
X=pd.DataFrame(X)
y=pd.Series(y)

from tqdm.notebook import tqdm, trange

for i in tqdm(range(0,len(X))):
    xi = np.asarray([X.iloc[i]])
    yi = y[i]
    preds['y'].append(yi)
    #do our simple models
    for name,model in river_models.items():
        pred = model.predict(xi)[0]
        if type(pred) is np.ndarray:
            pred = pred[0]
        for metric_k,metric_v in metrics.items():
            score = metric_v[name].update(yi, pred).get()
            scores[metric_k][name].append(score)
        preds[name].append(pred)
    #add in pls-deep
    x_t = pls_scaler.transform(xi)
    pred = pls_deep_model(torch.Tensor(x_t)).detach().numpy()[0]
    preds['pls_deep'].append(pred)
    for metric_k,metric_v in metrics.items():
        score = metric_v['pls_deep'].update(yi, pred).get()
        scores[metric_k]['pls_deep'].append(score)
    #add in pls-deep-lwr
    x_tt = pls_deep_model.compress(torch.Tensor(x_t)).detach().numpy()
    pred = pls_deep_lwr.predict(x_tt)[0]
    preds['pls_deep_lwr'].append(pred)
    for metric_k,metric_v in metrics.items():
        score = metric_v['pls_deep_lwr'].update(yi, pred).get()
        scores[metric_k]['pls_deep_lwr'].append(score)
    #add in deep
    x_t = deep_scaler.transform(xi)
    pred = deep_model(torch.Tensor(x_t)).detach().numpy()[0]
    preds['deep'].append(pred)
    
    for metric_k,metric_v in metrics.items():
        score = metric_v['deep'].update(yi, pred).get()
        scores[metric_k]['deep'].append(score)
        
    #add in deep-lwr
    x_t = deep_scaler.transform(xi)
    x_tt = deep_model.compress(torch.Tensor(x_t)).detach().numpy()
    pred = deep_lwr.predict(x_tt)[0]
    preds['deep_lwr'].append(pred)
    for metric_k,metric_v in metrics.items():
        score = metric_v['deep_lwr'].update(yi, pred).get()
        scores[metric_k]['deep_lwr'].append(score)
```

```python
scores_df = pd.DataFrame(preds)

def delist(l):
    l = np.asarray(l)
    for i,_ in enumerate(l):
        if type(l[i]) is list:
            l[i] = l[i][0]
    return l

for (columnName, columnData) in scores_df.iteritems():
    if not columnName is y:
        scores_df[columnName+"_dif"] = np.absolute(np.subtract(delist(columnData),np.asarray(scores_df["y"].tolist())))
scores_df.to_csv(log_dir/"preds.csv")
```
